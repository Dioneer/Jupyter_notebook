**Условие 1: Задача 1**

1.1. Скачать следующие данные: kc-house-data и laptop_price

```python
file_path = "kc_house_data.csv"
```

```python
df = pd.read_csv(file_path )
```

```python
df.head(7)
```

```python
def change(x):
    if x == 0:
        return "have waterfront"
    else:
        return "haven't waterfront"

df["waterfront_view"] = df["waterfront"].apply(change)
```

1.2. Изучите стоимости недвижимости

```python
plt.figure(figsize=(6,4))
plt.title("Price of realt")
plt.xlabel("price")
plt.ylabel("count")
plt.hist(df["price"], bins=35);
```

вывод: количество домой прямопропорционально их ценеене


1.3. Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(6,4))
plt.title("Proportion of living space")
plt.xlabel("sqft_living")
plt.ylabel("count")
plt.hist(df["sqft_living"], bins=15);
```

вывод:
Средний объем жилой площади превалирует

1.4. Изучите распределение года постройки

```python
plt.figure(figsize=(6,4))
plt.title("Proportion year of built")
plt.xlabel("sqft_living")
plt.ylabel("count")
plt.hist(df["yr_built"], bins=15);
```

вывод: большая часть домов достаточно новые

**Условие 3: 3 задача**

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
plt.figure(figsize=(10, 8))

corrMatrix = round(df.drop(
    columns=['id', 'date', 'zipcode', 'lat', 'long', 'sqft_living15', 'sqft_lot15', "waterfront_view"]
).corr(method='spearman'), 2)
corrMatrix[abs(corrMatrix) < 0.2] = 0

sns.heatmap(corrMatrix, annot=True, linewidths=.5, cmap='coolwarm')
plt.title("Матрица корреляции признаков");

```

```python
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='reg')
plt.title("Цена относительно жилой площади")
plt.xlabel("sqft_living")
plt.ylabel("price");
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(x=df["bedrooms"], y=df["price"],whis=1.5, showfliers=False)
plt.title("Цена относительно спален")
plt.xlabel("Price")
plt.ylabel("bedrooms");
```

```python
 grouped = df.groupby('sqft_lot')
```
