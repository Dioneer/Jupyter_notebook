```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('laptop_price.csv',sep=',', encoding='windows-1251')
```

```python
def change(x):
    if x[-2:] == "kg":
        return x[:-2]
    else:
        return x
df["Weightt"] = df["Weight"].apply(change)

df.head()
```

```python
def separate(x):
    x = x.split(" ")[1:]
    x = ' '.join(x)
    return x
    

df["Memory_Type"]=df["Memory"].apply(separate)
df.head()
```

```python
def separate(x):
    res1 = x.split(" ")[0]
    if res1[-2:] == "TB":
        x = float(res1[:-2])*1000
    else:
        x = float(res1[:-2])
    return x

df["Memory_Amount"]=df["Memory"].apply(separate)
df.head()
```

```python
def separate(x):
    return x.split(" ")[0]
    

df["Cpu_Company"]=df["Cpu"].apply(separate)
df.head()
```

```python
df["Weightt"]= df["Weightt"].astype(float)
```

```python
data5 = pd.crosstab(index=df["Company"],columns=df["Cpu_Company"])
data5 = data5.drop(columns=["Samsung"])
data5
```

```python
plt.figure(figsize=(10,8))
ticks = np.arange(len(data5["AMD"]))
offset1=0.2
col_width=0.4
plt.bar(ticks - offset1, data5["AMD"], width=col_width)
plt.bar(ticks + offset1, data5["Intel"], width=col_width)
plt.title("Memory_Type")
plt.xlabel("Name")
plt.ylabel("count")
plt.xticks(ticks, data5.index, rotation=30);
```

```python
data6 = pd.crosstab(index=df["Company"],columns=df["Cpu_Company"], normalize='index')
data6 = data6.drop(columns=["Samsung"]).reset_index()
data6
```

```python
plt.figure(figsize=(10,8))
plt.bar(data6["Company"], data6["AMD"])
plt.bar(data6["Company"], data6["Intel"], bottom=data6["AMD"] )
plt.title("Memory_Type")
plt.xlabel("Name")
plt.ylabel("count")
plt.xticks(rotation=30);
```
