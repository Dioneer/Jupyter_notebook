```python
import pandas as pd
```

```python
fale_path = 'laptop_price.csv'
```

```python
df = pd.read_csv(fale_path, sep=',', encoding='windows-1251')
df.head()
```

1.1 Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU

```python
def separate(x):
    return x.split(" ")[0]
    

df["Cpu_Company"]=df["Cpu"].apply(separate)
df.head()
```

1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя

```python
def separate(x):
    res1 = x.split(" ")[0]
    if res1[-2:] == "TB":
        x = float(res1[:-2])*1000
    else:
        x = float(res1[:-2])
    return x

df["Memory_Amount"]=df["Memory"].apply(separate)
df.head()
```

1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)

```python
def separate(x):
    x = x.split(" ")[1:]
    x = ' '.join(x)
    return x
    

df["Memory_Type"]=df["Memory"].apply(separate)
df.head()
```

1.4 Удалите признаки Memory и ScreenResolution

```python
def separate(x):
    res1 = x.split(" ")[0]
    if res1[-2:] == "TB":
        return True
    else:
        return False


df1 = df["Memory"].apply(separate)
```

```python
df[df1]
```

```python
df["Memory"] = 0
df["ScreenResolution"] = 0
df.drop(columns=["Memory", "ScreenResolution"],inplace=True)
df.head()
```

2.1 Создайте признак SSD, который изначально равен 0

```python
df["SSD"] = 0
df.head()
```

```python
def change(x):
    if x == "SSD":
        return 1
    else:
        return 0

df["SSD"] = df["Memory_Type"].apply(change)
df.head()
```

2.3 Уберите в признаке Weight значения 'kg' и поменяйте его тип данных на вещественный

```python
def change(x):
    if x[-2:] == "kg":
        return x[:-2]
    else:
        return x
df["Weight"] = df["Weight"].apply(change)

df.head()
```
